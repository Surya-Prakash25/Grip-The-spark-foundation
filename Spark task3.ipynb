```python
#importing the required libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings ('ignore')


#load the datasets into pandas dataframes

df = pd.read_csv("SampleSuperstore.csv")
print(df)
```

```python
#DATA VISUALIZATION USING PAIRPLOT

plt.figure(figsize=(8,5))
sns.kdeplot(df['Sales'],color='red',label='Sales',shade=True)
sns.kdeplot(df['Profit'],color='Blue',label='Profit',shade=True)
plt.xlim([-100,1000])
plt.legend()
```

```python
# Heatmap showing correlation between sales , profit and quantity

sns.heatmap(df[["Sales" , "Profit" , "Quantity"]].corr(), annot = True)
plt.title("Correlation between Sales , Profit and Quantity")
plt.show()
```

```python
# Bar chart showing profit margin by shipping mode

sns.barplot(x='Ship Mode',y='Profit',data=df)
plt.title("Profit Margin by Shipping Mode")
plt.show()
```

```python
#scatter plot showing sales and profit by Region

sns.scatterplot(x='Sales',y='Profit',hue='Region',data=df)
plt.title("Sales and Profit by Region")
plt.show()
```
